To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
from openai import OpenAI
from dotenv import load_dotenv
import os
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GEMINI = 'gemini-3-flash-preview'
MODEL_LLAMA = 'llama3.2'
GEMINI_BASE_URL = 'https://generativelanguage.googleapis.com/v1beta/openai/'
```

```python
# set up environment

load_dotenv(override=True)
api_key = os.getenv("GEMINI_API_KEY")

gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=api_key)
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

system_prompt = "You are a tech tutor. Explain the topics and answer the questions like you are teaching a beginner. Use examples or sample to explain if necessary."

message = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": question}
]
```

```python
# Get gpt-4o-mini to answer, with streaming
stream = gemini.chat.completions.create(model=MODEL_GEMINI, messages=message, stream=True)

response = ""
display_handle = display(Markdown(""), display_id=True)

for chunk in stream:
    response += chunk.choices[0].delta.content or ""
    update_display(Markdown(response), display_id=display_handle.display_id)

```

```python
# Get Llama 3.2 to answer
res = ollama.chat.completions.create(model='llama3.2', messages=message)

display(Markdown(res.choices[0].message.content))
```
